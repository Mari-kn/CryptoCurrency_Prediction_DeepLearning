# Multivariate Single Target Multi Step Time Series

```python
TRAIN_SPLIT=int(len(df)*0.8)
df=df[["high","number_of_trades"]]
dataset=np.array(df)
```

```python
# To Supervised
def multivariate_data(dataset, target, start_index, end_index, history_size,
                      target_size, step, single_step=False):
    data = []
    labels = []
    start_index = start_index + history_size
    if end_index is None:
        end_index = len(dataset) - target_size

    for i in range(start_index, end_index):
        indices = range(i-history_size, i, step)
        data.append(dataset[indices])
        if single_step:
            labels.append(target[i+target_size])
        else:
            labels.append(target[i:i+target_size])
    return np.array(data), np.array(labels)
```

```python
lag=1
past_history = lag  #we used number of lag data to predict
STEP = 1
n_step=30 #we_wanna_predict_30_days_later
x_train_multi, y_train_multi= multivariate_data(dataset, dataset[:,0], 0, #[:,0] 0 refers to high column in dataset
                                                TRAIN_SPLIT, past_history,
                                                n_step, STEP,
                                                single_step=False)
x_test_multi, y_test_multi = multivariate_data(dataset, dataset[:, 0],   #[:,0] 0 refers to high column in dataset
                                               TRAIN_SPLIT, None, past_history,
                                               n_step, STEP,
                                               single_step=False)
```

```python
print("(Num of observations, lag, number of features)")
print("x_train_multi_shape: ",x_train_multi.shape)
print("x_test_multi_shape: ", x_test_multi.shape)
```

```python
print("(Num of observations, n_step)")
print("y_train_multi_shape: ",y_train_multi.shape)
print("y_test_multi_shape: ",y_test_multi.shape)
```

```python
#Scale
from sklearn.preprocessing import MinMaxScaler
scalers = {}
for i in range(x_train_multi.shape[1]):
    scalers [i]= MinMaxScaler(feature_range=(0,1))
    x_train_multi[:, i, :] = scalers[i].fit_transform(x_train_multi[:, i, :])

for i in range(x_test_multi.shape[1]):
    x_test_multi[:, i, :] = scalers[i].transform(x_test_multi[:, i, :])

x_train_multi.shape,x_test_multi.shape
```

```python
scalers
```

```python
scaler_y =MinMaxScaler(feature_range=(0,1))
y_train_multi=scaler_y.fit_transform(y_train_multi)
y_test_multi=scaler_y.transform(y_test_multi)

y_test_multi.shape,y_train_multi.shape
```

#SIMPLE RNN

```python
from keras.models import Sequential
from keras.layers import LSTM,Dense,GRU,SimpleRNN,Conv1D,MaxPooling1D,Flatten,Dropout,BatchNormalization
RNN_model = Sequential()
RNN_model.add(SimpleRNN(50,input_shape=x_train_multi.shape[-2:],activation='relu',return_sequences=True))
#RNN_model.add(Dropout(0.2))
#RNN_model.add(BatchNormalization())
RNN_model.add(SimpleRNN(50,activation='relu',kernel_regularizer='l2'))
#RNN_model.add(BatchNormalization())
#RNN_model.add(Dropout(0.2))
RNN_model.add(Dense(n_step,activation='selu'))
RNN_model.compile(optimizer='adam', loss='mse')
```

```python
history_RNN=RNN_model.fit(x_train_multi,y_train_multi,
                          epochs=100,batch_size=16,validation_data=(x_test_multi,y_test_multi))
```

# Learning *Curve*

```python
plt.figure(figsize=(12,8))
plt.plot(history_RNN.history['loss'])
plt.plot(history_RNN.history['val_loss'])
plt.title('Model Loss for RNN')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
#Predict Test Set
predict_RNN=RNN_model.predict(x_test_multi)
predict_RNN.shape
```

```python
#MSE
from sklearn.metrics import mean_squared_error
mse_RNN= mean_squared_error(y_test_multi, predict_RNN)
mse_RNN
```
